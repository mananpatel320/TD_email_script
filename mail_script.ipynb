```python
def get_invgs(course_name):
    dictn = {
        "Course_ID" : [],
        "Course_Name" : [],
        "Exam_Date" : [],
        "Start_Time" : [],
        "End_Time"  : [],
        "Name" : [],
        "E_mail" : []
    }
    for i in range(len(inv_emails)):
        if(course_name == inv_course_names[i]):
            dictn["Course_ID"].append(inv_course_nums[i])
            dictn["Course_Name"].append(inv_course_names[i])
            dictn["Exam_Date"].append(str(inv_exam_dates[i])[0:11])
            dictn["Start_Time"].append(str(inv_exam_start[i]))
            dictn["End_Time"].append(str(inv_exam_end[i]))
            dictn["Name"].append(inv_names[i])
            dictn["E_mail"].append(inv_emails[i])
    data = pd.DataFrame(dictn)
    return data
```
