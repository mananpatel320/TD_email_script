```python
import xlrd
import time
import pandas as pd
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from pretty_html_table import build_table
```

```python
pathInv = "/home/mananpatel/Documents/BITS_Pilani_Hyderabad_Campus/TD/INVIGI.xlsx"
openFileInv = pd.read_excel(pathInv, engine='openpyxl', sheet_name='Sheet1').dropna()
```

```python
inv_course_nums = openFileInv['COURSE NO.']
inv_course_names = openFileInv['COURSE TITILE']
inv_exam_dates = openFileInv['DATE']
inv_exam_start = openFileInv['START TIME']
inv_exam_end = openFileInv['END TIME']
ic_names = openFileInv['IC NAME']
inv_names = openFileInv['INVIGILATOR NAME']
inv_emails = openFileInv['MAIL ID']
inv_id = openFileInv['X']
print(inv_id, inv_course_nums, inv_names, inv_course_names, inv_exam_dates, inv_exam_start, inv_exam_end, ic_names, inv_emails)
```

```python
email = "td@hyderabad.bits-pilani.ac.in"
password = ""
```

```python
i = 0
while i < len(inv_names):
    dictn = {
        "Course ID" : [],
        "Course Name" : [],
        "Exam Date" : [],
        "Start Time" : [],
        "End Time"  : [],
        "IC Name" : [],
    }
    j = i
    while (j < len(inv_names) and inv_names[i] == inv_names[j]):
        dictn["Course ID"].append(inv_course_nums[j])
        dictn["Course Name"].append(inv_course_names[j])
        dictn["Exam Date"].append(str(inv_exam_dates[j])[0:11])
        dictn["Start Time"].append(str(inv_exam_start[j]))
        dictn["End Time"].append(str(inv_exam_end[j]))
        dictn["IC Name"].append(ic_names[j])
        j = j + 1
    data = pd.DataFrame(dictn)
    print(data)
    i = j
```
